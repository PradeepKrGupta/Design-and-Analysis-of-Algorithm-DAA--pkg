```python
# importing the packages
import matplotlib.pyplot as plt
import timeit
import random
```

```python
# Program for Selection Sort
def selection(a, s):
    for step in range(s):
        min_idx = step
        for i in range(step + 1, s):
            if a[i] < a[min_idx]:
                min_idx = i
        (a[step], a[min_idx]) = (a[min_idx], a[step])

# Initialize lists to store sizes and corresponding execution times
sizes = []
execution_times = []

# Perform sorting for different input sizes and record execution times
for size in range(100, 1001, 100):
    data = [random.randint(0, 10000) for _ in range(size)]
    sizes.append(size)
    
    start = timeit.default_timer()
    selection(data, size)
    end = timeit.default_timer()
    
    execution_time_ms = (end - start) * 1000
    execution_times.append(execution_time_ms)
print(f"Time taken :{execution_time_ms}")

# Plot the results
plt.plot(sizes, execution_times, marker='o')
plt.title('Selection Sort Execution Time vs. Input Size')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (ms)')
plt.grid(True)
plt.show()
```

```python
# Insertion Sort

def insertion(a):
    for step in range(1, len(a)):
        key = a[step]
        j = step - 1
        while j >= 0 and key < a[j]:
            a[j + 1] = a[j]
            j = j - 1
        a[j + 1] = key

# Initialize lists to store sizes and corresponding execution times
sizes = []
execution_times = []

# Perform sorting for different input sizes and record execution times
for size in range(100, 1001, 100):
    data = [random.randint(0, 10000) for _ in range(size)]
    sizes.append(size)
    
    start = timeit.default_timer()
    insertion(data)
    end = timeit.default_timer()
    
    execution_time_ms = (end - start) * 1000
    execution_times.append(execution_time_ms)

print(f"Time taken :{execution_time_ms}")
# Plot the results
plt.plot(sizes, execution_times, marker='o')
plt.title('Insertion Sort Execution Time vs. Input Size')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (ms)')
plt.grid(True)
plt.show()
```

```python
# Bubble Sorting

def bubble(arr):
    for i in range(len(arr)):
        swapped = False
        for j in range(0, len(arr) - i - 1):
            if arr[j] > arr[j + 1]:
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp
                swapped = True
        if not swapped:
            break

# Initialize lists to store sizes and corresponding execution times
sizes = []
execution_times = []

# Perform sorting for different input sizes and record execution times
for size in range(100, 1001, 100):
    data = [random.randint(0, 10000) for _ in range(size)]
    sizes.append(size)

    start = timeit.default_timer()
    bubble(data)
    end = timeit.default_timer()

    execution_time_ms = (end - start) * 1000
    execution_times.append(execution_time_ms)
    
print(f"Time taken :{execution_time_ms}")
# Plot the results
plt.plot(sizes, execution_times, marker='o')
plt.title('Bubble Sort Execution Time vs. Input Size')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (ms)')
plt.grid(True)
plt.show()
```

```python
# Combining all the sorting algorithm and plot their time complexity in same graph
import random
import timeit
import matplotlib.pyplot as plt

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

# Initialize lists to store sizes and corresponding execution times for each algorithm
sizes = []
selection_times = []
insertion_times = []
bubble_times = []

# Perform sorting for different input sizes and record execution times for each algorithm
for size in range(100, 1001, 100):
    data = [random.randint(0, 10000) for _ in range(size)]
    sizes.append(size)

    # Measure execution time for selection sort
    start = timeit.default_timer()
    selection_sort(data.copy())
    end = timeit.default_timer()
    selection_times.append((end - start) * 1000)

    # Measure execution time for insertion sort
    start = timeit.default_timer()
    insertion_sort(data.copy())
    end = timeit.default_timer()
    insertion_times.append((end - start) * 1000)

    # Measure execution time for bubble sort
    start = timeit.default_timer()
    bubble_sort(data.copy())
    end = timeit.default_timer()
    bubble_times.append((end - start) * 1000)


# Time taken by each of them
print(f"Time taken by Selection sort :{selection_times}")
print(f"Time taken by Insertion sort :{insertion_times}")
print(f"Time taken by Bubble sort :{bubble_times}")

# Plot the results for all three sorting algorithms
plt.plot(sizes, selection_times, label='Selection Sort', marker='o')
plt.plot(sizes, insertion_times, label='Insertion Sort', marker='o')
plt.plot(sizes, bubble_times, label='Bubble Sort', marker='o')

plt.title('Sorting Algorithm Comparison')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid(True)
plt.show()
```
